```python
import pandas as pd

# 定义文件路径
file_path = 'by_stock/000001.SZ.csv'

# 使用 pandas 读取 CSV 文件
try:
    # 尝试读取文件
    # parse_dates 参数尝试将指定列（通常是第一列）解析为日期
    # index_col 将日期列设为索引
    df = pd.read_csv(file_path)
    
    # 按日期升序排列（通常股票数据源可能是降序的）
    df.sort_index(inplace=True)
    
    print("文件加载成功！")
    print("-" * 30)
    print(f"数据形状 (行数, 列数): {df.shape}")
    print("-" * 30)
    print("前 5 行数据:")
    print(df.head())
    print("-" * 30)
    print("数据类型:")
    print(df.dtypes)
    
except FileNotFoundError:
    print(f"错误：未找到文件 {file_path}，请检查路径是否正确。")
except Exception as e:
    print(f"发生错误: {e}")
```

```python
import os
import pandas as pd

# 输入输出目录
input_dir = 'by_stock'
output_dir = 'new'

# 如果输出目录不存在则创建
os.makedirs(output_dir, exist_ok=True)

# 遍历所有 csv 文件
for file_name in os.listdir(input_dir):
    if file_name.endswith('.csv'):
        file_path = os.path.join(input_dir, file_name)
        
        try:
            df = pd.read_csv(file_path)
            
            # 转换日期列为 datetime
            df['交易日期'] = pd.to_datetime(df['交易日期'])
            
            # 筛选 2024 年数据
            df_2024 = df[df['交易日期'].dt.year == 2024]
            
            # 如果有数据才保存
            if not df_2024.empty:
                output_path = os.path.join(output_dir, file_name)
                df_2024.to_csv(output_path, index=False)
                print(f"{file_name} 处理完成，保存至 {output_path}")
            else:
                print(f"{file_name} 在 2024 年没有数据")
        
        except Exception as e:
            print(f"{file_name} 处理失败: {e}")

print("全部处理完成！")
```
